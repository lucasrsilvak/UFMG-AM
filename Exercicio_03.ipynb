```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = load_data('Default')

df['default'] = df['default'].apply(lambda x: 1 if x == 'Yes' else 0)
df['student'] = df['student'].apply(lambda x: 1 if x == 'Yes' else 0)

X = df[['balance', 'income', 'student']]
y = df['default']

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train[['balance', 'income']] = scaler.fit_transform(X_train[['balance', 'income']])
X_test[['balance', 'income']] = scaler.transform(X_test[['balance', 'income']])
```

```python

model_balance = LogisticRegression()
model_balance.fit(X_train[['balance']], y_train)
y_pred_balance = model_balance.predict(X_test[['balance']])

print("Resultados com apenas 'balance':")
print("Acurácia:", accuracy_score(y_test, y_pred_balance))
print(confusion_matrix(y_test, y_pred_balance))
print(classification_report(y_test, y_pred_balance))

coef = model_balance.coef_[0]
intercept = model_balance.intercept_[0]
features = X.columns

print("\nCoeficientes do modelo:")
print(coef)
print(intercept)

model_student = LogisticRegression()
model_student.fit(X_train[['student']], y_train)
y_pred_student = model_student.predict(X_test[['student']])

print("\nResultados com apenas 'student':")
print("Acurácia:", accuracy_score(y_test, y_pred_student))
print(confusion_matrix(y_test, y_pred_student))
print(classification_report(y_test, y_pred_student))

coef = model_student.coef_[0]
intercept = model_student.intercept_[0]
features = X.columns

print("\nCoeficientes do modelo:")
print(coef)
print(intercept)

model_income = LogisticRegression()
model_income.fit(X_train[['income']], y_train)
y_pred_income = model_income.predict(X_test[['income']])

print("\nResultados com apenas 'income':")
print("Acurácia:", accuracy_score(y_test, y_pred_income))
print(confusion_matrix(y_test, y_pred_income))
print(classification_report(y_test, y_pred_income))

coef = model_income.coef_[0]
intercept = model_income.intercept_[0]
features = X.columns

print("\nCoeficientes do modelo:")
print(coef)
print(intercept)

model_all = LogisticRegression()
model_all.fit(X_train, y_train)
y_pred_all = model_all.predict(X_test)

print("\nResultados com 'balance', 'student' e 'income':")
print("Acurácia:", accuracy_score(y_test, y_pred_all))
print(confusion_matrix(y_test, y_pred_all))
print(classification_report(y_test, y_pred_all))
```
